```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links_df = pd.read_csv('links.csv')
```

```python
sub_links = links_df['links']
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup

# Configure Chrome options
chrome_options = Options()


# Initialize the WebDriver
driver = webdriver.Chrome()

texts = []

for sub_link in sub_links:
    try:
        # Set a page load timeout
        driver.set_page_load_timeout(120)  # Adjust the timeout as needed
        driver.get(sub_link)

        # Get the rendered HTML
        html = driver.page_source

        # Parse the HTML with BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')

        # Find the target div with the specified class
        content = soup.find('div', class_="detail-content afcbc-body")

        # Check if the content exists before accessing it
        if content:
            # Extract all <h1> to <h6> and <p> tags
            headings_and_paragraphs = content.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p'])
            text = ""
            # Collect the text from each tag
            for tag in headings_and_paragraphs:
                text += tag.get_text(strip=True) + "\n"
                print(f"{tag.get_text(strip=True)}")
            texts.append(text)
        else:
            print(f"Target div not found in: {sub_link}")
            
    except TimeoutException:
        print(f"Timeout occurred while accessing: {sub_link}")
        texts.append("")
    except Exception as e:
        print(f"An error occurred with {sub_link}: {e}")
        texts.append("")

# Close the driver
driver.quit()
```
